```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
import scipy as sp
```

```python
x = np.linspace (0, 10, 200) 
list_a = np.linspace(1,10,10)
list_mean = []
list_var = []
for i in range(len(list_a)):
    y1 = sts.gamma.pdf(x, a = list_a[i], scale = 2.5/(list_a[i]))
    plt.plot(x,y1)
    mean=sp.integrate.trapz(y1 * x, x)
    list_mean = np.append(list_mean, sp.integrate.trapz(y1 * x, x))
    list_var = np.append(list_var, np.sum((y1 - mean)**2)/len(x))
    print (list_mean[i],list_var[i])
```

```python
import numpy as np
from scipy.optimize import minimize
def f(x):
    return (2.3 - x[0])**2 +(0.5 - x[1]/((x[2]-1)*x[3]))**2 + (2.75 - x[1]/(x[2] - 1))**2 + (1 - x[1]**2/((x[2]-1)**2*(x[2] - 2)))**2
for x_initial in [
np.array([-0,-0,-0,-0]),
np.array([-5,-5,-5,-5]),
np.array([5,5,5,5]),
np.array([4,4,4,4])
]:
    result = minimize(f, x_initial)
    x_final = result.x
    print('Started at', x_initial)
    print('Ended at', x_final)
    print('f(%s) = %.6f' % (x_final, f(x_final)))
    print()
```
